```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
from nltk.util import ngrams
```

```python
# Sample text corpus (you can replace this with your actual text data)
text_corpus = df['comment']

# # Download the stopwords list (you only need to do this once)
# nltk.download('stopwords')
# nltk.download('punkt')

# Function to tokenize and preprocess each comment
def preprocess_comment(comment):
    words = word_tokenize(comment)
    words = [word.lower() for word in words if word.isalnum()]
    return words

# Tokenize and preprocess each comment in the DataFrame
tokenized_comments = text_corpus.apply(preprocess_comment)

# Flatten the list of tokenized words from all comments
words = [word for comment in tokenized_comments for word in comment]

# Get the set of Arabic stopwords
stop_words = set(stopwords.words('arabic'))

# Filter out stopwords from the list of words
filtered_words = [word for word in words if word not in stop_words]

# Calculate the frequency of each word using Counter
word_frequency = Counter(filtered_words)

# Sort the word frequency dictionary in descending order
sorted_word_frequency = dict(sorted(word_frequency.items(), key=lambda item: item[1], reverse=True))

# Print the word frequency results
print("Word Frequency Analysis:")
for word, frequency in sorted_word_frequency.items():
    print(f"{word}: {frequency}")
```

```python
def preprocess_comment(comment):
    words = nltk.word_tokenize(comment)
    words = [word.lower() for word in words if word.isalnum()]
    return words

# Tokenize and preprocess each comment in the DataFrame
df['tokenized_comment'] = df['comment'].apply(preprocess_comment)

# Get the set of Arabic stopwords
stop_words = set(stopwords.words('arabic'))

# Function to remove stopwords from a list of words
def remove_stopwords(words):
    return [word for word in words if word not in stop_words]

# Remove stopwords from the tokenized comments
df['cleaned_comment'] = df['tokenized_comment'].apply(remove_stopwords)
```

```python
# Function to extract n-grams from a list of words
def extract_ngrams(words, n):
    return list(ngrams(words, n))

# Define the value of n for n-grams
n = 3  # Change this value to get different n-grams (e.g., n=3 for trigrams)

# # Function to filter out NaN values from a list of words
# def remove_nan_words(words):
#     return [word for word in words if isinstance(word, str)]

# # Filter out NaN values from the 'cleaned_comment' column
# df['cleaned_comment'] = df['cleaned_comment'].apply(remove_nan_words)

# # Function to remove empty lists from the 'cleaned_comment' column
# def remove_empty_lists(words):
#     return words if words else None

# # Remove empty lists (resulting from NaN values) from the 'cleaned_comment' column
# df['cleaned_comment'] = df['cleaned_comment'].apply(remove_empty_lists)

# # Drop rows with NaN values in the 'cleaned_comment' column
# df.dropna(subset=['cleaned_comment'], inplace=True)

# Extract n-grams from the cleaned comments and create a list of all n-grams
all_ngrams = df['cleaned_comment'].apply(lambda words: extract_ngrams(words, n)).sum()

# Calculate the frequency of each n-gram in the entire corpus
ngram_frequency = Counter(all_ngrams)

# Print the top frequent n-grams in the entire corpus
print(f"Top {n}-grams in the entire corpus:")
for ngram, frequency in ngram_frequency.most_common(10):
    print(f"{ngram}: {frequency}")
```
